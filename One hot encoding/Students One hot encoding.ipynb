```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
file = pd.read_csv("Datasets/Students Marks.csv")
```

```python
file
```

```python
dummies = pd.get_dummies(file.School)
```

```python
dummies
```

```python
file = pd.concat([file, dummies], axis = "columns")
```

```python
file
```

```python
file = file.drop(['School', 'PQR School'], axis = 'columns')
```

```python
file
```

```python
X = file.drop(['Average'], axis = 'columns')
```

```python
X
```

```python
Y = file.Average
```

```python
Y
```

```python
model = LinearRegression()
```

```python
model.fit(X, Y)
```

```python
model.predict([[85, 90, 88, 1, 0]])
```

```python
model.score(X, Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
import pickle
```

```python
with open('Marks Model', 'wb') as f:
    pickle.dump(model,f)
```

```python
with open('Marks Model', 'rb') as f:
    modelFile = pickle.load(f)
```

```python
modelFile.predict([[85, 90, 88, 1, 0]])
```

```python
file = pd.read_csv("Datasets/HomePrice.csv")
```

```python
file
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
file.town = label.fit_transform(file.town)
```

```python
file
```

```python
X = file[['town', 'area']]
```

```python
X
```

```python
X = X.values
```

```python
X
```

```python
Y = file[['price']].values
```

```python
Y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```
