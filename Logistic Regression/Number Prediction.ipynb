```python
from sklearn.model_selection import train_test_split
```

```python
train_X, test_X, train_y, test_y = train_test_split(file.data, file.target, test_size=0.2, random_state = 10)
```

```python
len(train_X)
```

```python
len(test_X)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(train_X, train_y)
```

```python
model.score(test_X, test_y)
```

```python
file.target[100]
```

```python
model.predict([file.data[100]])
```

```python
y_predicted = model.predict(test_X)
y_predicted
```

```python
test_y
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(test_y, y_predicted)
```

```python
cm
```

```python
import seaborn as sn 
plt.figure(figsize = (10,8))  # plt.figure(figsize= (10, 8)) figure of size 10 X 8 inches is drawn
sn.heatmap(cm, annot=True)  # annot = True adds all the numerical value of the cm into the heatmap
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
