```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
file = pd.read_csv("Datasets/Canada Income.csv")
```

```python
regObject = linear_model.LinearRegression()
```

```python
regObject.fit(file[["year"]], file.income)
```

```python
file.head()
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Income")
plt.scatter(file.year, file.income)
```

```python
%matplotlib inline 
plt.xlabel("Year")
plt.ylabel("Income")
plt.scatter(file.year, file.income)
plt.plot(file.year, regObject.predict(file[["year"]]))
```

```python
y2020 = regObject.predict([[2020]])
```
