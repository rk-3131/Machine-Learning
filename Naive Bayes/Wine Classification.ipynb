```python
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
dir(wine)
```

```python
import pandas as pd
```

```python
file = pd.DataFrame(wine.data, columns = wine.feature_names)
```

```python
file.head()
```

```python
file['Type'] = wine.target
```

```python
file.head()
```

```python
file['Type_name'] = file.Type.apply(lambda x : wine.target_names[x])
```

```python
file.head()
```

```python
temp = file.drop('Type_name', axis = 'columns')
cor_matrix = temp.corr()
```

```python
cor_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10, 7))
sns.heatmap(cor_matrix, annot=True)
```

# The value of the Type depends mostly on
1. total_phenols
2. flavanoids
3. od280/od315_of_diluted_wines

```python
X = file[['total_phenols', 'flavanoids', 'od280/od315_of_diluted_wines']]
```

```python
X.head()
```

```python
Y = file.Type
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.2)
```

```python
len(train_X)
```

```python
len(test_X)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
g_model = GaussianNB()
```

```python
g_model.fit(train_X, train_Y)
```

```python
score_gaussian = g_model.score(test_X, test_Y)
```

```python
score_gaussian
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
m_model = MultinomialNB()
```

```python
m_model.fit(train_X, train_Y)
```

```python
score_multiNB = m_model.score(test_X, test_Y)
```

```python
score_multiNB
```

```python
# Gaussian model is better than that of the multiNB
```
