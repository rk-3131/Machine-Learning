```python
file['ExerciseAngina'] = file.ExerciseAngina.apply(lambda x : 1 if x == "Y" else 0)
```

```python
file.head()
```

```python
file.isna().sum()
```

```python
ECG_dummies = pd.get_dummies(file.RestingECG)
```

```python
chest_pain_dummies = pd.get_dummies(file.ChestPainType)
```

```python
file['Sex'] = file.Sex.apply(lambda x : 1 if x == 'M' else 0)
```

```python
ST_Slope_dummies = pd.get_dummies(file.ST_Slope)
```

```python
file = pd.concat([ECG_dummies, chest_pain_dummies, ST_Slope_dummies])
```
